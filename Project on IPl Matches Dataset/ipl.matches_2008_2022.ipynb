```python
# find the last match palyed by kohli in delhi
ipl.head()
```

```python
ipl['all_palyers'] = ipl['Team1Players']+ipl['Team2Players']
ipl.head()
```

```python
def did_kohli_played(player_list):
    return 'V kohli' in player_list
```

```python
ipl['did_kohli_play'] = ipl['all_palyers'].apply(did_kohli_played)
ipl[(ipl['City'] == 'Delhi') & (ipl['did_kohli_play'] == True)].drop_duplicates(subset=['City', 'did_kohli_play'],keep='first')
```
