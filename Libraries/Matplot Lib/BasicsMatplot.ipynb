# Bar Chart

• Bivariate Analysis

• Use case - Aggregate analysis of groups

```python
# bar chart

children = [10,20,40,10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt. bar(colors, children, color='black')  # vertical graph
```

```python
# horrizontal bar chart : when the x-axis pe agar category bafh jata hai tb..
plt.barh(colors, children, color = 'black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

bar chart : side by side

```python
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, color = 'red')
plt.bar(np.arange(df.shape[0]),df['2016'], width=0.2, color = 'blue')
plt.bar(np.arange(df.shape[0]) +0.2, df['2017'], width=0.2, color='brown')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem

children = [10,20,40,10,30]
colors = ['red red red red red red', 'blue blue blue blue', 'green green green green green', 'yellow yellow yellow yellow ', 'pink pink pink pink']
plt. bar(colors, children, color='black')
plt.xticks(rotation='vertical') # jitna name bda hoga utna sheet ko spread kar dega..
```

```python
# stacked abr chart
plt.bar(df['batsman'], df['2017'], label='2017', color = 'black')
plt.bar(df['batsman'], df['2016'],bottom=df['2017'], label='2016', color = 'blue')
plt.bar(df['batsman'], df['2015'],bottom=df['2016']+df['2017'], label='2015', color = 'red')

plt.legend() 
plt.show()
```

# Pie Chart



100% me kiska kitna contribution hai...

Univariate / biavariate analysis

categorical vs numerical

use case - to find contribution on a standard scale

```python
# simple data
data = [23,45, 100, 20, 49]
subjects = ['eng', 'science', 'maths', 'hindi', 'sst']
plt.pie(data, labels=subjects, autopct='%1.1f%%')
plt.show()
```
