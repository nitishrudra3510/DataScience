# 3D surface plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x, y)
xx.shape
```

```python
z = xx**2+yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx, yy, z, cmap='viridis') # give the 3 d surface

fig.colorbar(p)

plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx, yy, z, cmap='viridis')
plt.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx, yy, z, cmap='viridis')
plt.colorbar(p)

plt.show()
```

# contour plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
xx.shape

z = xx**2+yy**2
z.shape
```

3D graph to convert in 2D graph

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contour(xx, yy, z, cmap='viridis') 

fig.colorbar(p)

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contourf(xx, yy, z, cmap='viridis') 

fig.colorbar(p)

plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx, yy, z, cmap='viridis')
plt.colorbar(p)
```

# pivot table : 

Parameters
		data  DataFrame jisme se pivot table banani hai

		values  Wo column jiske upar aggregation apply karna hai

		index  Rows ke liye grouping ka column

		columns  Columns ke liye grouping ka column
    
		aggfunc  Aggregation function jaise sum, mean, count (default: mean)

```python
# on multiple dataframes
# using tips 
tips.pivot_table(index=['day', 'time'], columns=['sex', 'smoker'], values='total_bill', aggfunc='mean').plot(kind='line')
```

```python
tips.pivot_table(index=['day', 'time'], columns=['sex', 'smoker'], values='total_bill', aggfunc='mean').plot(kind='bar')
```

```python
tips.pivot_table(index=['day', 'time'], columns=['sex', 'smoker'], values='total_bill', aggfunc='mean').plot(kind='pie', subplots=True, figsize=(20, 20))
plt.show()
```
