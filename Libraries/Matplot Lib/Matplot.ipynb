# Basic Plots


# Line plots:
there are useful for visualizing the relationship between
two variables over a continuous interval. To create a line plot,
use the plot() function in Matplotlib.

plt.xlabel('X-axis') sets the label for the x-axis to "X-axis".

plt.ylabel('Y-axis') sets the label for the y-axis to "Y-axis".

plt.title('Bar plot') sets the title of the plot to "graph plot".

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.xlabel('X-axis') # x-axis

plt.ylabel('Y-axis')

plt.title('Line Plot')

plt.show() # display the plot on the screen..
```

# Scatter Plots:
Scatter plots are ideal for displaying the distribution and
relationship between two numerical variables. Matplotlib
provides the scatter()

```python

x = [1,2,3,4,5]
y = [2,4,6,8, 10]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')

plt.show()
```

# Bar Plots:
It is effective for comapring categorical data or showing the distribution of a single variable. Matplotlib offers the bar() and barh()
functions for craeting vertical and horizontal bar plots.

```python
categories = ['A', 'B', "C", "D"]
values = [10, 20, 15, 5]
plt.bar(categories, values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar plot')

plt.show()
```

# Histograms:
It is useful for visualizing the distribution of continous variable. It provides the hist() function to create histograms..

```python
import numpy as np
data = np.random.randn(1000) # generate the data

plt.hist(data, bins=30) # bins: the range of data should be divided into 30 equal-width intervals, or bins.
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram plot')

plt.show()
```

# Legends:
It can specify the psoition of the legend and add labels for each data series..

using legends()

```python

x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

# Plotting two series
plt.plot(x1, y1, label='Series 1')  # Plot Series 1
plt.plot(x2, y2, label='Series 2')  # Plot Series 2

# Adding legend with specified location
plt.legend(loc='upper right')  # Place legend in the upper right corner

# Display the plot
plt.show()
```

# $.5 Axis Limits and Ticks..
it can sets the limits and ticks for the x-axis ad y-axis using the xlim(), ylim(), and xticks(), yticks() functions..


```python
import matplotlib.pyplot as plt

plt.plot(x1, y1)  # Plot the data points defined by x1 and y1
plt.xlim(0, 10)   # Set the x-axis limits from 0 to 10
plt.ylim(0, 20)   # Set the y-axis limits from 0 to 20
plt.xticks([0, 5, 10])  # Set custom tick marks on the x-axis at 0, 5, and 10
plt.yticks([0, 10, 20])  # Set custom tick marks on the y-axis at 0, 10, and 20

plt.xlabel('X-axis')  # Add label to the x-axis
plt.ylabel('Y-axis')  # Add label to the y-axis
plt.title('Plot with Custom Axis Limits and Ticks')  # Add plot title

plt.grid(True)  # Show grid lines on the plot

plt.show()  # Display the plot
```
