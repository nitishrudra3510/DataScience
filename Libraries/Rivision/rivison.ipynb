What is the average difference in price between smartphones with fast charging capabilities and those without?

```python
fast_charging_supported = df[df['fast_charging_available']==True]['price'].mean()
```

```python
fast_charging_not_supported = df[df['fast_charging_available']==False]['price'].mean()
```

```python
difference = fast_charging_supported-fast_charging_not_supported
difference
```
