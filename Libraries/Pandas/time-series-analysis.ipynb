# Timedelta Object

represents a duration, the difference between two dates or times..

```python
# creating using timestamp objects
t1 = pd.Timestamp('6th january 2023 8:20:14')
t2 = pd.Timestamp('26th january 2023 10:00:00')
```

```python
t2-t1
```

```python
# standalone creation

pd.Timedelta(days=2, hours=10, minutes=10)
```

```python
# Arithmetic 
pd.Timestamp('6th jan 2023') + pd.Timedelta(days=2, hours=10, minutes=10)
```

```python
pd.date_range(start='2023-01-06',end='2023-1-31', freq='D') + pd.Timedelta(days=2, hours=10, minutes=10)
```

time data ke sath add substract bhi kar sakte hai..

```python
pd.date_range(start='2023-01-06',end='2023-1-31', freq='D') - pd.Timedelta(days=2, hours=10, minutes=10)
```

# Resampling : 


```python
# downsampling q karna chahiye

yt['subscribers'].plot() 
yt['subscribers'].resample('w').mean().plot() # yeha par daily data ko weekly data me convert kar diya and mean fucntion apply kar diya
```

```python
# unsampling : 

google['Close'].resample('12H').interpolate(method='spline', order=2).plot()
```

```python
google['Close'].resample('M').interpolate(method='spline', order=2).plot()
```

# Rolling Window(Smoothing)

```python
# rolling window
yt['subscribers'].plot()
yt['subscribers'].rolling(3).mean().plot() # 3 means : in inmage
```

```python
# yt['subscribers'].plot(title='original')
yt['subscribers'].rolling(3).mean().plot(title='rolling')
yt['subscribers'].ewm(3).mean().plot(title='ewm')
```
