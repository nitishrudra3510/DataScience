# Timedelta Object

represents a duration, the difference between two dates or times..

```python
# creating using timestamp objects
t1 = pd.Timestamp('6th january 2023 8:20:14')
t2 = pd.Timestamp('26th january 2023 10:00:00')
```

```python
t2-t1
```

```python
# standalone creation

pd.Timedelta(days=2, hours=10, minutes=10)
```

```python
# Arithmetic 
pd.Timestamp('6th jan 2023') + pd.Timedelta(days=2, hours=10, minutes=10)
```

```python
pd.date_range(start='2023-01-06',end='2023-1-31', freq='D') + pd.Timedelta(days=2, hours=10, minutes=10)
```

time data ke sath add substract bhi kar sakte hai..

```python
pd.date_range(start='2023-01-06',end='2023-1-31', freq='D') - pd.Timedelta(days=2, hours=10, minutes=10)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Load the dataset
df = pd.read_csv("youtube_dataset.csv")

# Generate 100 sequential dates starting from today
start_date = datetime.today()
df["date"] = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(len(df))]

# Save the updated dataset
df.to_csv("youtube_dataset.csv", index=False)
```
