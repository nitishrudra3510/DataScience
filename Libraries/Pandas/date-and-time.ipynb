# DateTimeIndex object

A collection of pandas timestamp

DatetimeIndex : it suppports multiple date
timestamp : supports one date only

```python
# from string
pd.DatetimeIndex(['2023/1/1', '2024/1/1', '2023/1/2'])
```

```python
# using python datatime object..
pd.DatetimeIndex([dt.datetime(2023, 1, 1), dt.datetime(2022, 1, 1), dt.datetime(2021, 1, 1)])
```

```python
# using pd.timestamps..

dt_index = pd.DatetimeIndex([pd.Timestamp(2023, 1, 1), pd.Timestamp(2022, 1, 1), pd.Timestamp(2021, 1, 1)])
```

```python
# using datetimeindex as series index

pd.Series([1,2,3], index=dt_index)
```

date_range function

```python
# generate daily dates in a given range..

pd.date_range(start='2023/1/5', end='2023/2/28', freq='2D')
```

```python
# business days..
pd.date_range(start='2023/1/5', end='2023/2/28', freq='B')
```

```python
# w -> one week per day..
pd.date_range(start='2023/1/5', end='2023/2/28', freq='W-THU')
```

```python
# H -> hourly date(factor)

pd.date_range(start='2023/1/5', end='2023/2/28', freq='H')
```

per 5 hours

```python
pd.date_range(start='2023/1/5', end='2023/2/28', freq='6H')
```

```python
# M-> month end:

pd.date_range(start='2023/1/5', end='2023/2/28', freq='M')
```

```python
# MS -> month start

pd.date_range(start='2023/1/5', end='2023/2/28', freq='MS')
```

```python
# A -> year end
pd.date_range(start='2023/1/5', end='2023/2/28', freq='A')
```

```python
# using periods (number of results)
pd.date_range(start='2023/1/5', periods=25, freq='D')
```

```python
pd.date_range(start='2023/1/5', periods=35, freq='6H')
```

```python
pd.date_range(start='2023/1/5', periods=25, freq='M')
```

# to_datetime function


converts an existing objects to pandas timestamp/datetimeindex object..

```python
# simple series example
s = pd.Series(['2023-01-05', '2023-02-06', '2024-01-07'])
```

```python
pd.to_datetime(s)
```

```python
pd.to_datetime(s).dt.year
```

```python
# with errors
s = pd.Series(['2023-01-05', '2023-02-06', '2024-130-07'])
pd.to_datetime(s)
```

```python
pd.to_datetime(s, errors='coerce') # code crase kardiye hai tb errors = 'coerce' use karte hai..
```
