aggregate method : specify kakrta hai ki konse columns par hm konse aggregate function lagta hai..

```python
# passing dictionary..
# genres.sum() 

genres.agg(
    {
        "Runtime" : "mean",
        "IMDB_Rating"  : "mean",
        "No_of_Votes" : "sum",
        "Gross" : "sum",
        "Metascore" : "min"
        
    }
)
```

```python
# passing list : every gcolumn ke under alag se three column bna gya hai min, amx, and sum
genres.agg(['min', 'max', 'sum'])
```

```python
# looping in the group
for group, data in genres:
    print(data)
```

```python
for group, data in genres:
    print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
```

add the highest rated movies value in the dataframe

```python
df = pd.DataFrame(columns=movies.columns)

for group, data in genres:
    max_rating_data = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, max_rating_data], ignore_index=True)
```

```python
# find ranking of each movie in the group according to IMDB score..
```

```python
def rank_movie(group):
    group['genres_rank'] =  genres['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
# find normalized IMDBB rating group wise 
```

```python
def normal(group):
    group['norm_rating'] = ((group["IMDB_Rating"] - group['IMDB_Rating'].min())/(group["IMDB_Rating"].max() - group['IMDB_Rating'].min()))
    return group
genres.apply(normal)
```
