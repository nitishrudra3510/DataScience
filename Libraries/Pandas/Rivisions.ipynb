To calculate the length of the array generated by np.arange(2.0, 2.6, 0.1), you can use the following formula:
length = ceil((stop - start) / step)

Let's break down the calculation step-by-step:
Start: The starting value is 2.0.
Stop: The ending value is 2.6 (not included).
Step: The step size is 0.1.
Now, let's plug these values into the formula:
length = ceil((2.6 - 2.0) / 0.1)

Calculating the numerator: 2.6 - 2.0 = 0.6
Dividing the numerator by the step size: 0.6 / 0.1 = 6.0
Using the ceil() function to round up the result to the nearest integer: ceil(6.0) = 7
Therefore, the length of the array generated by np.arange(2.0, 2.6, 0.1) is 7

# Groupby()

It is a powerfull function that allow you to group data based on one or more leys (columns)
and perform aggregate operations on those. It's particullary useful for data analysis and manipulation enabling you to summarize or transform your data in meaningful waways

Key Features of groupby

	1.	Grouping Data: You can group data by one or more columns, which creates a new object that can be used for further operations.
	2.	Aggregation: After grouping, you can perform aggregation functions such as sum(), mean(), count(), min(), and max() on the grouped data to get summary statistics.
	3.	Transformation: You can also transform the data, applying a function to each group without reducing the number of rows.
	4.	Filtering: You can filter groups based on specific conditions.

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Values': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

grouped = df.groupby('Category')

result = grouped['Values'].sum()
result
```

Example Explained

		Creating a DataFrame: The example creates a DataFrame with two columns: Category and Values.
		Grouping: The groupby('Category') groups the data by the values in the Category column.
		Aggregation: The sum() function calculates the total for the Values column within each category.

# Merging

1. Merging (pd.merge())

Merging combines two dataframes based on a key or multiple keys (similar to SQL joins).

```python
import pandas as pd

# Create two dataframes
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'age': [25, 30, 35]
})
merged_df = pd.merge(df1, df2, on='id')
merged_df
```

In this case, the two dataframes are merged on the id column. Only matching rows (by default, an inner join) are returned.

# Concatenating (pd.concat())

Concatenating stacks dataframes either vertically (row-wise) or horizontally (column-wise).


```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'age': [25, 30]
})

df2 = pd.DataFrame({
    'name': ['Charlie', 'David'],
    'age': [35, 40]
})

concatinate_df = pd.concat([df1, df2])
concatinate_df
```

# join()

Joining combines two dataframes on their index, but columns can be specified as well.

```python
# Create two dataframes
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'age': [25, 30]
}, index=[1, 2])

df2 = pd.DataFrame({
    'city': ['New York', 'San Francisco']
}, index=[1, 2])


# joined based on the index

joined_df = df1.join(df2)

joined_df
```

# Time Series()

it is referes to a sequence of data points indexed by time, often recorded at consistent iontervals(eg: daily, monthly, yearly). Time series data is common in many fields like finnance, economics, environmental studies, etc like where tracking how a variable changes over time.

Key Functions for Time Series:

pd.to_datetime(): Converts strings or numeric data into datetime objects.

df.set_index(): Sets a column (such as a datetime column) as the index for time-based operations.

resample(): Aggregates data by a new frequency (e.g., convert daily data to monthly data).

shift(): Shifts the data points forward or backward by a specific number of periods, useful for lagging or leading data.

```python
# creating time series:

dates = pd.date_range('2024-01-01', periods=10, freq='D')

data = [120, 130, 125, 140, 150, 155, 144, 160, 135, 170]

df = pd.DataFrame({"Date": dates, "value": data})
df

df.set_index('Date', inplace=True)

df
```

```python
# resampling the data
# resample the data to a different frequency, like weekly ('W'), monthly ('M'), or yearly ('Y').
weekly_df = df.resample('w').mean()
weekly_df
```

```python
# Time Series Shifting (LAgging/ Leading)
# shift() allows you to move data forward or backward. This is useful for creating lag features in time series analysis.

# Shift the data one period back (lagging)
df['Value_Lagged'] = df['value'].shift(1)
print(df)

# Shift the data one period forward (leading)
df['Value_Lead'] = df['value'].shift(-1)
print(df)
```

```python
import pandas as pd

# Creating a time series DataFrame
dates = pd.date_range('2024-01-01', periods=5, freq='D')
data = [100, 150, 130, 170, 160]
df = pd.DataFrame({'Date': dates, 'Value': data})

# Setting Date as the index
df.set_index('Date', inplace=True)


print(df)

# Resampling the data to weekly frequency
weekly_df = df.resample('W').mean()
print(weekly_df)
```

```python
# Rolling Windows(Moving Average):

# 3-dau moving average

df['Moving_Avg'] = df['value'].rolling(window=3).mean().round(2)
df
```

# Time Zone Conversion

can convert your time series to a specific time zone using tz_localize() and tz_convert().

```python


# Create a time series
rng = pd.date_range('2023-10-14', periods=5, freq='H')
df = pd.DataFrame({'value': [1, 2, 3, 4, 5]}, index=rng)

df
```

```python
# Localize to UTC
df_utc = df.tz_localize('UTC')

# Convert to US/Eastern
df_eastern = df_utc.tz_convert('US/Eastern')

print(df_eastern)
```

The time is now 4 hours earlier because US/Eastern is UTC-4 during daylight saving time (fall season).

```python
# Converting between time span representations
# Time span: it refers to the length or duration between two points in time.
#  typically represents the difference between two dates or timestamps, often measured in units like seconds, minutes, hours, days, weeks, or even years. 


rng = pd.date_range('1/1/2012', periods=5, freq='M')

ts = pd.Series(np.random.randn(len(rng)), index=rng)


print(ts)

# Convert the time series to a PeriodIndex, which represents periods (months) instead of timestamps
ps = ts.to_period()

# Print the PeriodIndex (ps)
print(ps)

# Convert the PeriodIndex back to a TimestampIndex, which returns to the original datetime format
ts_converted = ps.to_timestamp()

# Print the converted time series
print(ts_converted)
```

```python
# All operation of time series:

import pandas as pd
import numpy as np

# 1. Creating a Time Series
dates = pd.date_range('2024-01-01', periods=10, freq='D')
data = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190]
df = pd.DataFrame({'Date': dates, 'Value': data})

# Set 'Date' as the index
df.set_index('Date', inplace=True)
print("1. Created Time Series:\n", df)

# 2. Resampling Time Series (Daily to Weekly)
weekly_df = df.resample('W').mean()
print("\n2. Resampled Time Series (Weekly):\n", weekly_df)

# 3. Time Series Shifting (Lagging)
df['Value_Lagged'] = df['Value'].shift(1)
print("\n3. Shifted Time Series (Lagged):\n", df)

# 4. Rolling Windows (3-Day Moving Average)
df['3-Day_MA'] = df['Value'].rolling(window=3).mean()
print("\n4. 3-Day Moving Average:\n", df)

# 5. Time Series Slicing (Extracting data between specific dates)
sliced_df = df['2024-01-03':'2024-01-07']
print("\n5. Sliced Time Series (2024-01-03 to 2024-01-07):\n", sliced_df)

# 6. Date Range Generation (Hourly frequency)
hourly_dates = pd.date_range('2024-01-01', periods=10, freq='H')
hourly_df = pd.DataFrame({'Date': hourly_dates, 'Value': np.arange(10)})
hourly_df.set_index('Date', inplace=True)
print("\n6. Hourly Time Series:\n", hourly_df)

# 7. Handling Missing Data (Forward Fill)
df_missing = df.asfreq('D').ffill()
print("\n7. Forward Filled Missing Data:\n", df_missing)

# 8. Time Series Aggregation (Monthly Sum)
monthly_aggregation = df.resample('M').sum()
print("\n8. Monthly Aggregated Time Series (Sum):\n", monthly_aggregation)

# 9. Time Zone Conversion (UTC to US/Eastern)
df_tz = df.tz_localize('UTC').tz_convert('US/Eastern')
print("\n9. Time Zone Converted Time Series:\n", df_tz)

# 10. Datetime Features Extraction (Year, Month, Day)
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
print("\n10. Datetime Features Extracted:\n", df[['Year', 'Month', 'Day']])


```

Converting between the period and timestamp enables some convenient arithmetic fucntions to be used. In the folowing example, we convert a quartely frequncy with year ending in November to 9 am of the end of the month following the quarter end():

```python
prng = pd.period_range('1990Q1', '2000Q4', freq='Q-NOV')
ts = pd.Series(np.random.randn(len(prng)), prng)

# Set the index of the time series to the end of each quarter, shifted by one month, and then to hourly frequency starting from 9 AM

ts.index = (prng.asfreq('M', 'e')+1).asfreq('H', 's')+9
ts.head
```
