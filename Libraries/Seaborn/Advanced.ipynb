# Regression Plots

• regplot

• Implot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips, x = 'total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x = 'total_bill', y='tip', hue='sex')
```

```python
# reside plot : error of predictions..

sns.residplot(data=tips, x = 'total_bill', y='tip')
```

plotting pairwise relatiosnhip (pairGrid vs pairplot)

```python
sns.pairplot(iris, hue='species')
```

```python
# pairgrid() : 

g = sns.PairGrid(data=iris, hue='species')

# g.map()
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.barplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.barplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.barplot)
g.map_offdiag(sns.histplot)
```

```python
# map_diag -> map_upper -> map_lower

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# var 


g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

# JointGrid vs JointPlot
