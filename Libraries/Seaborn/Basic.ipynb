# Basics Plotting Functions:

scatter plot : visualize the relationship between two numerical variables..

```python
# Scatter Plot 
sns.scatterplot(data=tips, x = "total_bill", y="tip", hue="time")
plt.show()

# to use the (data = tips) - to use the specific tips data
# (x = "total_bill") - it represent the values of total_bill on the x-axis...
# y="tip" - it represent the values of tips on the y- axis
# hue="time" - it will point of the color (dinner or lunch).
```

# bar plot 
comnapre the categorical data

```python
sns.barplot(data=tips, x = "day", y="total_bill", hue="sex")
plt.show()
```

# Histogram

display the distribution of a variables.

```python
sns.histplot(data=tips, x = "total_bill", bins=20, kde=True)
plt.show()
```

# Line Plot

track the data across a sequence or time..

```python
sns.lineplot(data=tips, x = "size", y="total_bill", ci=None)
plt.show
```

# InterMediate Plot

BoxPlot

display the distribution and outliers in data..

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
```

Violoin Plot

combine box plot and KDE to show data distribtuion..

```python
sns.violinplot(data=tips, x = "day", y = "total_bill", hue="sex", split=True)
plt.show()
```

# Pair Plot 

```python
sns.pairplot(data=tips, hue="sex")
```

# heatMap : 
visualize the matrices and co-relations..

```python
nunmeric_tips = tips.select_dtypes(include=["float", "int64"])
correlation_matrix = nunmeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

# Distribution Plot

used for univariate analysis

• used to find out the distribution

• Range of the observation

• Central Tendency

• is the data bimodal?

• Are there outliers?

# Plots under distribution plot

• histplot

• kdeplot

• rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# hist plot -> create bins(10-20)(20-30), etc

sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=20) # bins size reduce..
```


It's vizualize the distribution of a categorical variable using the logic of a histogram.

Discrete Dins are automatically set for categorical variables

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex', element='step')
```

```python
# faceting using col and row -> not work on hsitplot fucntion
sns.displot(data=tips, x='tip', kind='hist',col='sex', element='step')
```

```python
# kdeplot():

# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

sns.kdeplot(data=tips, x='total_bill')
```

```python
# hue -> fill
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

```python
# ?Rugplot


# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill


sns.histplot(data=tips, x='total_bill',y='tip')
```

```python
# Bivariate kdeplot

# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(data=tips, x='total_bill',y='tip')
```
