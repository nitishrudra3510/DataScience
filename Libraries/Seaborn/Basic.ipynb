```python
# import the library
import seaborn as sns
import matplotlib.pyplot as plt # often used to aloneside seaborns
```

# Built in Datasets

```python
tips = sns.load_dataset("tips")
tips.head()
```

# Basics Plotting Functions:

scatter plot : visualize the relationship between two numerical variables..

```python
# Scatter Plot 
sns.scatterplot(data=tips, x = "total_bill", y="tip", hue="time")
plt.show()

# to use the (data = tips) - to use the specific tips data
# (x = "total_bill") - it represent the values of total_bill on the x-axis...
# y="tip" - it represent the values of tips on the y- axis
# hue="time" - it will point of the color (dinner or lunch).
```

# bar plot 
comnapre the categorical data

```python
sns.barplot(data=tips, x = "day", y="total_bill", hue="sex")
plt.show()
```

# Histogram

display the distribution of a variables.

```python
sns.histplot(data=tips, x = "total_bill", bins=20, kde=True)
plt.show()
```

# Line Plot

track the data across a sequence or time..

```python
sns.lineplot(data=tips, x = "size", y="total_bill", ci=None)
plt.show
```

# InterMediate Plot

BoxPlot

display the distribution and outliers in data..

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
```

Violoin Plot

combine box plot and KDE to show data distribtuion..

```python
sns.violinplot(data=tips, x = "day", y = "total_bill", hue="sex", split=True)
plt.show()
```

# Pair Plot 

```python
sns.pairplot(data=tips, hue="sex")
```

# heatMap : 
visualize the matrices and co-relations..

```python
nunmeric_tips = tips.select_dtypes(include=["float", "int64"])
correlation_matrix = nunmeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
