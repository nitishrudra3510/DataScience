# Analyzing Stock Levels

Calculate Total Stock Quantity per Product

```python
total_stock = df.groupby('ProductName')['TotalItemQuantity'].sum().reset_index()
total_stock
```

plot total stock quantity per products

```python
plt.figure(figsize=(10,6))

sns.barplot(data=total_stock, x='ProductName', y='TotalItemQuantity', palette='viridis')
plt.title("Total stock quantity per Product")
plt.xticks(rotation=45)
plt.show()
```

# Analyzing Sales..

# total sales Amount per Product.

```python
total_sales = df.groupby('ProductName')['SalesAmount'].sum().reset_index()
total_sales
```

# Plot total sales amount per product


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=total_sales, x='ProductName', y='SalesAmount', palette='viridis')
plt.title('Total Sales Amount per Product')
plt.xticks(rotation=45)
plt.show()
```

# Analyzing Orders

# plot order status Distribution

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Status', hue = 'SalesOrderStatus', palette='viridis')
plt.title('Order Status Distribution')
plt.show()
```

# Sales Order Analysis


# Calculate total sales amount per order status


```python
sales_order_status = df.groupby('SalesOrderStatus')['SalesAmount'].sum().reset_index()
sales_order_status
```

# total sales amount per order status

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=sales_order_status, x='SalesOrderStatus', y='SalesAmount',hue = 'SalesOrderStatus', palette='viridis', dodge=False)
plt.title('Total Sales Amount per Order Status')
plt.show()
```

# Advanced Analysis with Numpy

Average profit per product.

```python
average_profit = df.groupby('CategoryName')['Profit'].mean().reset_index()
```

 Plot average profit per product category


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=average_profit, x='CategoryName', y='Profit', palette='viridis')
plt.title('Average Profit per Product Category')
plt.show()
```

# Correlation Analysis

correaltion Analysis

```python

correlation_matrix = df[['SalesAmount', 'ProductStandardCost', 'Profit', 'OrderItemQuantity', 'TotalItemQuantity']].corr()
correlation_matrix
```


# plot heatmap of the correlation matrix
