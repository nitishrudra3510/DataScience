# Acessing the data

loc[]:  access a group of rows and columns by labels or a boolean array.

```python
playstore.loc[2] # access rows with inddex label 2
```

```python
# accessing value at row index 2 and column 'A'
playstore.loc[2,'Installs']
```

```python
# acessing all the rows for Size's column
playstore.loc[:, ['Size']]
```

```python
playstore.loc[10:20, :]
```

```python
playstore.loc[2:5, 'Category':'Content Rating']  # Access rows from index 2 to 5 and columns from 'A' to 'C'
```

# iloc[]: Access a group of rows and columns by integer position(s).

```python
# acessing row at the index of the postion 2
playstore.iloc[2]
```

by integer slices, acting similar to the numpy/python


```python
playstore.iloc[2:5, 0:2]
```

```python
playstore.iloc[:, [0]] # prints all rows for 1st columns
```

```python
playstore.iloc[3, 4] # return the values 3 rows position and the 4th columns posiitons
```
