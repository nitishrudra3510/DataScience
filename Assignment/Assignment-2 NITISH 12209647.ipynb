Q1. In a 50 over match, for the first N overs Indian cricket team scored with a run rate of R that is R runs per over. Display the run rate with which Indian team should score a run in the remaining overs so that the final score of the team reaches to S. Values of N, R, and S should be entered by the user. Display the output upto two decimal places. Examples:

image.png

```python
def cal_required_run_rate(N, R, S):
  total_overs = 50
  runs_scored = N*R

  # remaining runs to be scored to reach the target S
  remain_runs = S - runs_scored

  # remaining over
  remaining_overs = total_overs-N

  # req run_rate for the remainng overs
  req_run_rate = remain_runs/remaining_overs

   # Rounding to two decimal places manually
  req_run_rate_rounded = int(req_run_rate * 100 + 0.5) / 100.0

  return req_run_rate_rounded


# user inputs
N1, R1, S1 = 20, 4, 300
N2, R2, S2 = 25, 3, 300
N3, R3, S3 = 30, 6, 300

# Calculating and prints the req run rates
print(f"N1={N1}, R1={R1}, S1={S1}: {cal_required_run_rate(N1, R1, S1)}")
print(f"N2={N2}, R2={R2}, S2={S2}: {cal_required_run_rate(N2, R2, S2)}")
print(f"N3={N3}, R3={R3}, S3={S3}: {cal_required_run_rate(N3, R3, S3)}")

```

Q2. If a single taxi can carry 'A' number of passengers, then display the number of taxis that will be required to carry 'N' number of passengers. Also display the distribution of the passengers in the taxis. Values of A, and N should be entered by the user. Assume N will always be greater than A.

image.png

```python
def cal_taxis(A, N):

    # Cal the number of taxis
    full_taxis = N // A

    # Cal the number of remaining passengers
    remaining_passengers = N % A

    # Calculate the total number of taxis

    total_taxis_no = full_taxis + (1 if remaining_passengers > 0 else 0)

    # Display the results
    print("Number of Taxis:", total_taxis_no)
    if remaining_passengers > 0:
        print("Distribution of Passengers:", A * full_taxis, "passengers in", full_taxis, "taxis, and", remaining_passengers, "passenger" if remaining_passengers == 1 else "passengers", "in 1 taxi")
    else:
        print("Distribution of Passengers:", A * full_taxis, "passengers in", full_taxis, "taxis, and 0 passengers in 1 taxi")

# take the input from the users.
A = int(input("Enter the number of passengers a single taxi can carry A : "))
N = int(input("Enter the number of passengers N : "))

cal_taxis(A, N)




```

Q3. You are supposed to develop a simulation for a 2D game that involves characters moving around on a grid. Each character has a position represented by a Vector object with x and y coordinates. You need to implement movement and interaction features for these characters using operator overloading.
Requirements:
 1.	Implement a Vector class to represent the position of characters.
  2.	Overload the + operator to move a character by adding another vector to its position.
   3.	Overload the - operator to calculate the distance between two characters.
    4.	Overload the * operator to scale a character's movement by a scalar.
     5.	Overload the == operator to check if two characters are at the same position.
      6.	Overload the __str__ method to provide a string representation of a character's position.

```python
# Define a class Vector
class Vector:

    def __init__(self, x, y):
        self.x = x
        self.y = y


    # overload the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)


    def __sub__(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5



    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)


    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(5, 7)

print("Initial positions:")
print("v1:", v1)
print("v2:", v2)

# Moving v1 by v2
v1 = v1 + v2
print("\nAfter moving v1 by v2:")
print("v1:", v1)

# Calculating distance between v1 and v2
distance = v1 - v2
print("\nDistance between v1 and v2:", distance)

# Scaling v2 by 2
v2 = v2 * 2
print("\nAfter scaling v2 by 2:")
print("v2:", v2)

# Checking if v1 and v2 are at the same position
Position_equal = v1 == v2
print("\nAre v1 and v2 at the same position ?", Position_equal)
```
